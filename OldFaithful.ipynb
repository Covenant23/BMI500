```python
#import necessary documents

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

#read the data set
df = pd.read_csv('OldFaithfulDataset.csv', usecols=[i for i in range(2)])

#add data to a data frame
data = pd.DataFrame(df)

#extract values for line of best fit
x = df['eruptions'].values
y = df['waiting'].values

# found out if there are any outliers
print(data.min())
print(data.max())

#calculate line of bet fit
a,b = np.polyfit(x,y,1)

#Plot the scatter plot
plt.scatter(data['eruptions'], data['waiting'], s = 27 )

#plot line of best fit
plt.plot(x, a*x+b, color = 'r') 

plt.title("Old Faithful Eruptions vs Waiting Time")
#set axis labels
plt.xlabel('''Eruptions Time(mins)


                  Fig1.Shows the relationship between eruption duration and the time intervals between eruptions.''')
plt.ylabel('Waiting Time(mins)')

#add ledgend
plt.legend(["Data" , "Line of Best Fit"], bbox_to_anchor = (1 , 1))

plt.show()

#data
```
